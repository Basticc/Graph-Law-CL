A veces los ministerios aparecen con distitnos nombres, aveces con acentos, aveces con la omisión de  
una letra, por ejemplo:

```Ministerio de Justicia, Ministerio de Justicia y Seguridad, ...```

Por lo que **simplifique** la clasificación de ministerios en base a una palabra clave, si el texto  
del organismo contiene una palabra **clave** lo categorizo como cierto ministerio, por ejemplo, si  
contiene la palabra *justicia* guardo el nombre en un set, asi al final tengo set de todas las referencias  
hacia el ministerio de justicia.

Esto lo hago para los organismos que fui observando hasta clasificar a todos, basicante eso hace  
la siguiente celda de codigo.


**LIMITACIÓN**: Podria pasar que una ley pertenece a dos ministerio, por lo tanto, estoy simplificando  
el problema a que esto no pasa, podria arreglarlo en un futuro cercano y se añadan a ambos ministerios

````python
import json


with open("data/parsed_law.json", "r", encoding="utf-8") as f:
    law_chile = json.load(f)


for law_code in law_chile:

    law = law_chile[law_code]
    mod = law["organismo"]

    if "justicia" in mod:
        justicia.add(mod)

    elif "interior" in mod:
        interio.add(mod)

    elif "economia" in mod:
        economia.add(mod)

    elif "relaciones"  in mod:
        relaciones.add(mod)

    elif "vivienda"  in mod:
        vivienda.add(mod)

    elif "salud"  in mod or "salubridad" in mod:
        salud.add(mod)

    elif "mineria"  in mod:
        mineria.add(mod)

    elif "agricultura"  in mod:
        agricultura.add(mod)

    elif "trabajo" in mod:
        trabajo.add(mod)

    elif "educacion" in mod:
        educacion.add(mod)

    elif "bienes" in mod:
        bienes.add(mod)

    elif "transporte" in mod:
        transporte.add(mod)

    elif "obras" in mod:
        obras.add(mod)

    elif "secretaria" in mod:
        secretaria.add(mod)

    elif "defensa" in mod:
        defensa.add(mod)

    elif "planificacion" in mod:
        planificacion.add(mod)

    elif "tierras" in mod:
        tierras.add(mod)

    elif "fomento" in mod:
        fomento.add(mod)

    elif "energia" in mod:
        energia.add(mod)

    elif "social" in mod:
        social.add(mod)

    elif "ambiente" in mod:
        ambiente.add(mod)

    elif "culturas" in mod:
        cultura.add(mod)
    elif "no especificado" in mod:
        no_especifico.add(mod)

    elif "deporte" in mod:
        deporte.add(mod)

    elif "seguridad" in mod:
        seguridad.add(mod)

    elif "marina" in mod:
        marina.add(mod)

    elif "ciencia" in mod:
        ciencia.add(mod)

    elif "instruccion" in mod:
        instruccion.add(mod)

    elif "ferrocarril" in mod:
        ferrocarril.add(mod)

    elif "mujer" in mod:
        mujer.add(mod)
    elif "guerra" in mod:
        guerra.add(mod)
    elif len(mod)<2:
        no_especifico.add(mod)
    elif "hacienda" in mod:

        continue
    else:
        misc.add(mod)



org_dict = {
            "ministerio de justicia": list(justicia),
            "ministerio del interior": list(interio),
            "ministerio de economia": list(economia),
            "ministerio de relaciones": list(relaciones),
            "ministerio de vivienda": list(vivienda),
            "ministerio de salud": list(salud),
            "ministerio de trabajo": list(trabajo),
            "ministerio de agricultura": list(agricultura),
            "ministerio de educacion": list(educacion),
            "ministerio de bienes": list(bienes),
            "ministerio de transporte": list(transporte),
            "ministerio de obras": list(obras),
            "ministerio de secretaria": list(secretaria),
            "ministerio de defensa": list(defensa),
            "ministerio de planificacion": list(planificacion),
            "ministerio de tierras": list(tierras),
            "ministerio de fomento": list(fomento),
            "ministerio de energia": list(energia),
            "ministerio de social": list(social),
            "ministerio de ambiente": list(ambiente),
            "ministerio de cultura": list(cultura),
            "ministerio de deporte": list(deporte),
            "ministerio de seguridad": list(seguridad),
            "ministerio de marina": list(marina),
            "ministerio de mineria": list(mineria),
            "ministerio de ciencia": list(ciencia),
            "ministerio de instruccion": list(instruccion),
            "ministerio de ferrocarril": list(ferrocarril),
            "ministerio de mujer": list(mujer),
            "ministerio de guerra": list(guerra),
            "ministerio de no especifico": list(no_especifico),
            "ministerio de hacienda": ["ministerio de hacienda"],
            "ministerio de misc": list(misc)
            }
````

````python
import json

with open("data/parsed_law.json", "r", encoding="utf-8") as f:
    law_chile = json.load(f)

org_map = {}
for std_name, variants in org_dict.items():
    for v in variants:
        org_map[v] = std_name.upper()  

for code, law in law_chile.items():
    org = law.get("organismo", "").strip()

    # Si el organismo aparece en nuestro mapa, lo reemplazamos
    if org in org_map:
        law["organismo"] = org_map[org]


````

Guardamos los resultados

````python

with open("data/parsed_law.json", "w", encoding="utf-8") as f:
    json.dump(law_chile, f, ensure_ascii=False, indent=2)
````
